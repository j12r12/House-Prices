```python
a = [1,2,3,4,5]
b = [3,1,5,2,6]

c = np.dot(a,b)
c
```

```python
mag_x = np.sqrt(np.sum([i**2 for i in a]))
mag_y = np.sqrt(np.sum([j**2 for j in b]))

cos_theta = c / (mag_x * mag_y)
cos_theta
```

```python
theta = np.arccos(cos_theta)
theta
```

```python
from math import degrees

deg = degrees(theta)
```

```python
deg
```

```python
from scipy.stats import pearsonr
```

```python
pearsonr(a,b)
```

```python
d = [1,3,2.5,4.5,5]

p = pearsonr(a,d)

print(f"Correlation: {p[0]:.3f}\np-value: {p[1]:.3f}")
```

```python
# Now do it manually

def covariance(x,y):
    
    xbar = np.mean(x)
    ybar = np.mean(y)
    xys = [(i-xbar) * (j-ybar) for i,j in zip(x,y)]
    
    ans = (np.sum(xys)) / (len(x)-1)
    
    return ans
    
    
```

```python
covariance(a,b)
```

```python
covariance(a,d)
```

```python
# So all we can tell from covariance is that the slope is positive, but not the
# strength of the relationship as the scales may well be different.
```

```python
def correlation(x,y):

    xstd = np.sqrt(np.var(x, ddof=1))
    ystd = np.sqrt(np.var(y, ddof=1))
    
    corr = covariance(x,y) / (xstd*ystd)
    
    return corr
```

```python
correlation(a,b)
```

```python
correlation(a,d)
```

```python
pearsonr(a,d)
```

```python
nums = train.select_dtypes(exclude="object")
```

```python
ys = train["SalePrice"]
```

```python
nums.drop(["Id", "SalePrice"], inplace=True, axis=1)
```

```python
tups = [(pearsonr(nums[i],ys)) for i in ["ScreenPorch", "PoolArea"]]
```

```python
tups
```

```python
# So no linear correlation for ScreenPorch or PoolArea. Need to look for non-linear relationships.
```

```python
from scipy.stats import spearmanr
```

```python
tups_spear = [(spearmanr(nums[i],ys)) for i in ["ScreenPorch", "PoolArea"]]
```

```python
tups_spear
```

```python
# So still no correlation with these two and SalePrice
# How about with one another?
```
